Reading the data

```python
import pandas as pd
data=pd.read_csv('/content/EEG_data.csv.zip')
data
```

TARGET VARIABLE

```python
target=data['subject_understood']
target.head()
```

FEATURES

```python
x = data.drop(['subject_understood','video_id','subject_id'], axis=1)
x
```

```python
x.info()
```

```python
x.describe()
```

SPLITTING THE DATA

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,target,test_size=0.2,random_state=42)
```

NORMALIZATION

```python
x.isnull().sum()
```

```python
import pandas as pd
num_classes = target.nunique()
print("Number of classes:", num_classes)
```

```python
counts = target.value_counts()


count_0 = counts.get(0, 0)
count_1 = counts.get(1, 0)

print(f"Number of 0s: {count_0}")
print(f"Number of 1s: {count_1}")
```

PERCEPTRON

```python
from sklearn.linear_model import Perceptron
pm=Perceptron()
pm.fit(x_train,y_train)
```

```python
yp=pm.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
print(confusion_matrix(y_test,yp))
print(classification_report(y_test,yp))
print(accuracy_score(y_test,yp))
```
